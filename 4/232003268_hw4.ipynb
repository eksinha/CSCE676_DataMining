```python
congress_members_tweets = pd.read_json("us-congress-tweets/tweets-sample-400k.json", lines=True)
congress_members = pd.read_csv("us-congress-tweets/congress_members.csv")
congress_members_tweets = congress_members_tweets.dropna(subset=['text']).reset_index().drop('index', axis=1)
congress_members_tweets.head()
```

```python
mentioned_cg = []
mentioned_cg_ids = []
congress_members_tweets.entities[0]['user_mentions'][0]['id']
temp = congress_members.userid.tolist()
mentioned_cg = [congress_members_tweets.text[i] for i in range(len(congress_members_tweets.text)) if len(congress_members_tweets.entities[i]['user_mentions'])!=0 and congress_members_tweets.entities[i]['user_mentions'][0]['id'] in temp]
mentioned_cg_ids = [congress_members_tweets.id[i] for i in range(len(congress_members_tweets.text)) if len(congress_members_tweets.entities[i]['user_mentions'])!=0 and congress_members_tweets.entities[i]['user_mentions'][0]['id'] in temp]
```

```python
temp = congress_members.userid.tolist()
cg_tweet = [congress_members_tweets.text[i] for i in range(len(congress_members_tweets.text)) if congress_members_tweets.user[i]['id'] in temp]
cg_tweet_ids = [congress_members_tweets.id[i] for i in range(len(congress_members_tweets.text)) if congress_members_tweets.user[i]['id'] in temp]
```

```python
freq = congress_members_tweets.groupby('user_id').count().sort_values(['id'],ascending=False)['id'].reset_index()
freq.head()
```

```python
top_cg_tweeters = pd.merge(congress_members,freq, how='inner',left_on ='userid',right_on='user_id').reset_index().drop(['index','user_id'],axis=1).sort_values(['id'],ascending=False)
top_cg_tweeters.head()
```

```python
final_tweets = cg_tweet+mentioned_cg
final_tweets_ids = cg_tweet_ids+ mentioned_cg_ids
```

```python
stop_words = set(stopwords.words('english'))
words_to_keep = ['don', "don't", 'ain', 'aren', "aren't", 'couldn', "couldn't", 'didn', "didn't", 'doesn', "doesn't", 'hadn', "hadn't", 'hasn', "hasn't", 'haven', "haven't", 'isn', "isn't", 'ma', 'mightn', "mightn't", 'mustn', "mustn't", 'needn', "needn't", 'shan', "shan't", 'no', 'nor', 'not', 'shouldn', "shouldn't", 'wasn', "wasn't", 'weren', "weren't", 'won', "won't", 'wouldn', "wouldn't"]
my_stop_words = stop_words
for word in words_to_keep:
 my_stop_words.remove(word)
```

```python
tweet_dec = []
for i in final_tweets:
  if isinstance(i, str):
    tweet_dec.append(deaccent(i))
```

```python
countryList = dict(countries_for_language('en'))
countries_old = list(countryList.values())
countries = [i.lower() for i in countries_old]
```

```python
texts = [
    [word for word in document.lower().split() if (word not in my_stop_words and not re.match(r"http\S+", word) and re.match(r"[a-zA-Z]+|@+", word) and len(word)>2)]
    for document in tweet_dec
]
```

```python
model = Word2Vec(sentences=texts, window=20, min_count=1, negative=50, workers=8,vector_size=300)
model.save("word2vec.model")
```

```python
a = [model.wv.most_similar('@randpaul',topn=50)[p][0] for p in range(50) if (re.match("[^@]", model.wv.most_similar('@randpaul',topn=50)[p][0]))]
a
```

```python
congressList = [top_cg_tweeters.screen_name[i].lower() for i in range(21)]
```

```python
countriesList = defaultdict(list)
for i in countries:
  try:
    countriesList[i] = [model.wv.most_similar(i,topn=5)[p][0] for p in range(5) if(model.wv.most_similar(i.lower(),topn=5)[p][0] not in countries)]
  except:
    pass
```

```python
congressIn = defaultdict(list)
for p in congressList:
  try:
    # print(p)
    congressIn[p] = [model.wv.most_similar('@'+p,topn=5)[o][0] for o in range(5) if (re.match("[^@]", model.wv.most_similar('@'+p,topn=5)[o][0]))]
  except:
    pass
```

```python
final_arr = []
for i in congressIn.keys():
  temp = [model.wv.distance('@'+i, p) for p in countriesList.keys()]
  final_arr.append(temp)
```

```python
X_axis_label = list(countriesList.keys())
Y_axis_label = list(congressIn.keys())

plt.figure(figsize=(25,10))
cmap = sns.cm.rocket_r
ax=sns.heatmap(final_arr, cmap=cmap)
ax.set_xticklabels(X_axis_label)
ax.set_yticklabels(Y_axis_label)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")
plt.setp(ax.get_yticklabels(), rotation=0)
ax.set_title("Congressmen vs countries vector distance")
# # fig.tight_layout()
plt.show()
```
